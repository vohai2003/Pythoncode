Bài 5

```python
from random import randint
l = []
for i in range(0,1000):
  l.append(randint(-1000,1000))
fname = input("Nhap ten file:")
f = open(fname,"w")
temp = []
for i in range(0,10):
  temp.append(i)
for i in range(0,100):
  for j in range(0,10):
    temp[j] = str(l[i*10+j])
  f.write(",".join(temp)+"\n")
f.close()
f = open(fname,"r")
lines = f.readlines()
for i in range(0,100):
  lines[i] = "  ".join(lines[i].split(","))
  print(lines[i])
```
