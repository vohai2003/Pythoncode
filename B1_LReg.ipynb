```python
import numpy
```

```python
data = numpy.loadtxt("/content/gdrive/My Drive/Dataset/ex1data1.txt",delimiter=",")
```

```python
def data_extract(data):
  n = len(data)
  X = data[:,:-1]
  X = numpy.column_stack([numpy.ones((n),dtype=numpy.int32),X])
  y = data[:,-1:]
  y = numpy.reshape(y,(n,1))
  return X,y
```

```python
X,y = data_extract(data)
print(X)
print(y)
```
