```python
import os
import numpy
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.linear_model import LinearRegression
```

MinMaxScaler

```python
def minmaxfit(data):
  scaler = MinMaxScaler()
  scaler.fit(data) #fitting data
  X,y = data[:,:-1], data[:,-1]
  model = LinearRegression()
  model.fit(X,y)
  return model.coef_
```

```python
data = numpy.loadtxt("/content/gdrive/My Drive/Dataset/ex2data1.txt",delimiter=",")
print(data)
```

```python
w = minmaxfit(data)
print("Optimal Coef.:\n",w)
```

StandardScaler

```python
def standardfit(data):
  scaler = StandardScaler()
  scaler.fit(data) #fitting data
  data = scaler.transform(data) #scaling data
  X,y = data[:,:-1], data[:,-1]
  model = LinearRegression()
  model.fit(X,y)
  return model.coef_
```

```python
data = numpy.loadtxt("/content/gdrive/My Drive/Dataset/ex2data2.txt",delimiter=",")
print(data)
```

```python
w = standardfit(data)
print("Optimal Coef.:\n",w)
```
