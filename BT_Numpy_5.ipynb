1.3: Xuất ảnh ra màn hình

```python
from PIL import Image, ImageOps
im = Image.open(path+"sample.png")
img = im.convert("RGB")
display(img) #Trên máy tính sẽ dùng img.show()
```

1.4: Chuyển từ ảnh sang numpy array

```python
import numpy
img_array = numpy.array(img)
print("img_ndim:",img_array.ndim)
print("img_shape:",img_array.shape)
```

1.5: In nội dung của array thu được ra màn hình

```python
print(img_array)
```

1.6: Các thao tác nâng cao với ảnh

```python
#grayscale
img_gray = ImageOps.grayscale(img)
#resize ảnh
img_resize = img.resize((400,400),resample=5)
#lưu ảnh
img.save("img.png",fmt="png")
display(img_gray)
display(img_resize)
!ls
```

Bài 2: Thao tác với tensor (sử dụng Tensorflow)

```python
import tensorflow
dim1array = numpy.random.randint(1,100,size=(1))
dim2array = numpy.random.randint(1,100,size=(2,2))
dim3array = numpy.random.randint(1,100,size=(3,3,3))
dim4array = numpy.array([numpy.random.randint(1,100,size=(4,4,4)) for _ in range(0,4)])
dim1tensor = tensorflow.convert_to_tensor(dim1array)
dim2tensor = tensorflow.convert_to_tensor(dim2array)
dim3tensor = tensorflow.convert_to_tensor(dim3array)
dim4tensor = tensorflow.convert_to_tensor(dim4array)
print(dim1tensor)
print(dim2tensor)
print(dim3tensor)
print(dim4tensor)
```
